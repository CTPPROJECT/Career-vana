```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  # Corrected import statement
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import pygwalker as pyg

df = pd.read_csv('data/b_data.csv')

df.head(10)
```

```python
size_data = df.groupby(['agreeableness', 'salary', 'sector']).size().reset_index(name='count')

# Scatter plot with size
plt.figure(figsize=(12, 8))
sns.scatterplot(x='agreeableness', y='salary', hue='sector', size='count', sizes=(20, 200), data=size_data, palette='viridis')

# Customize plot
plt.title('Scatter Plot of Agreeableness vs Salary Range (Size Reflects Count)')
plt.xlabel('Agreeableness')
plt.ylabel('Salary Range')
plt.legend(title='Sector')

# Show the plot
plt.show()
```

```python
size_data = df.groupby(['agreeableness', 'salary', 'sector']).size().reset_index(name='count')

# Bar chart
plt.figure(figsize=(12, 8))
sns.barplot(x='agreeableness', y='count', hue='sector', data=size_data, palette='viridis')

# Customize plot
plt.title('Bar Chart of Agreeableness vs Count (Grouped by Salary and Sector)')
plt.xlabel('Agreeableness')
plt.ylabel('Count')
plt.legend(title='Sector', loc='upper right')

# Show the plot
plt.show()
```
